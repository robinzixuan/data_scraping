```python
import pylab
import psycopg2

def output():
    global cur
    conn = psycopg2.connect(database="", user="postgres", password="postgres", host="192.168.10.80", port="5432")
    cur = conn.cursor()
    cur.execute("CREATE TABLE result(id serial PRIMARY KEY,Name varchar(32),Status varchar(32),Description text,Refer text,Phase text,other text);")
    
    
def tablewrite(n,b,c,d,e,f):
    cur.execute("INSERT INTO test(Name,Status,Description,Refer,Phase,other)VALUES(%s, %s, %s, %s, %s, %s)", (n,b,c,d,e,f))
    
    
def plot_regression(x,y):
    '''Draws a regression line for values in lists x and y.
       x and y must be the same length.'''
    xarr = pylab.array(x) 
    yarr = pylab.array(y) 
    m,b = pylab.polyfit(xarr,yarr, deg = 1) 
    pylab.plot(xarr,m*xarr + b, '-') 
    
def plot(x_list,y_list,names):   
    '''Plot the values in the parameters.'''   
    pylab.xlabel('x')   #label x axis
    pylab.ylabel('y')   #label y axis   
    pylab.scatter(x_list,y_list)
    for i, txt in enumerate(names): 
        pylab.annotate(txt, (x_list[i],y_list[i]))    
    plot_regression('x','y')    
    pylab.show()                # displays the plot     
    
    
    
def analyze_file(n):   
    d=dict()
    file=open(n,'r')
    title = file.readline()
    for line in file:
        (num,user,time, x,y)=line.strip().split(",")
        if user not in d:
            d[user]=dict()
            d[user]["time"]=list()
            d[user]['tuple']=list()
            d[user]["time"].append(time)
            d[user]['tuple'].append((float(x),float(y)))
        else:
            d[user]['time'].append(time)
            d[user]['tuple'].append((float(x),float(y)))
    return d
    
def main():
    d1=analyze_file("checkin_dataset.csv")
    l1=list()
    x_list=list()
    y_list=list()
    for i in d1:
        l1.append(i)
        for q in d1[i]["tuple"]:
            x_list.append(q[0])
            y_list.append(q[1])

    plot(x_list,y_list,i)
    output()

main()
```
