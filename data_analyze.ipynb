```python
import pylab
def plot_regression(x,y):
    '''Draws a regression line for values in lists x and y.
       x and y must be the same length.'''
    xarr = pylab.array(x) #numpy array
    yarr = pylab.array(y) #numpy arry
    m,b = pylab.polyfit(xarr,yarr, deg = 1) #creates line, only takes numpy arrays
    #as parameters
    pylab.plot(xarr,m*xarr + b, '-') #plotting the regression line
    
 
def format_num(num):
   return "{:.2f}".format(num)    #define the format of number


def plot(x_list,y_list,names):   
    '''Plot the values in the parameters.'''   
    pylab.xlabel('x')   #label x axis
    pylab.ylabel('y')   #label y axis   
    pylab.scatter(x_list,y_list)
    for i, txt in enumerate(names): 
        pylab.annotate(txt, (x_list[i],y_list[i]))    
    plot_regression('x','y')    
    pylab.show()                # displays the plot     
    

def xy(d):
    d1=dict()
    for p in d:
        n=0
        summary=0
        for i in d[p]["tuple"]:
            tuple1=i
            x=tuple1[0]
            y=tuple1[1]
            result= x*y
            n+=1
            summary+result
        average=summary/n
        d1[p]=average
        average=format_num(average)
    return d1
            
        
    
def analyze_file(n):   
    d=dict()
    file=open(n,'r')
    title = file.readline()
    for line in file:
        (num,user,time, x,y)=line.strip().split(",")
        if user not in d:
            d[user]=dict()
            d[user]["time"]=list()
            d[user]['tuple']=list()
            d[user]["time"].append(time)
            d[user]['tuple'].append((float(x),float(y)))
        else:
            d[user]['time'].append(time)
            d[user]['tuple'].append((float(x),float(y)))
    return d
    
def main():
    d1=analyze_file("checkin_dataset.csv")
    d2=xy(d1)
    l1=list()
    x_list=list()
    y_list=list()
    a_list=list()
    b_list=list()
    for i in d1:
        l1.append(i)
        for q in d1[i]["tuple"]:
            x_list.append(q[0])
            y_list.append(q[1])
    plot(x_list,y_list,i)
    dict3=dict()
    for a in d2:
        if d2[a] not in  dict3:
            dict3[a]=1
        else:
            dict3[a]+=1
     maxium=0
     max_plot=0
     for k,v in dict3:
         if v>maxium:
             maxium=v
             max_plot=k
     print(max_plot)
     
    
    

main()


```
